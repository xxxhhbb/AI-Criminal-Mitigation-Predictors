```python
reduce=[]

pattern_ti1 = '，[^，]{0,10}建议[^，]*?减[^，]{0,10}刑[^月]{0,10}月'



def extract_sentence(text,pattern):
    # Check if the text is a string and is not empty
    if isinstance(text, str) and text:
        match = re.search(pattern, text)
        if match:
            return match.group(0)
    # Return an empty string if text is not a string or if no match is found
    return ' '

for id, tp in enumerate(texts):
    extracted_sentence = extract_sentence(tp, pattern_ti1)
    reduce.append(extracted_sentence)
# Apply the function to the second column of the dataframe and store the results

```

```python
import re

# 正则表达式模式
pattern = r'刑((一|二|两|三|四|五|六|七|八|九|十)年)?(一|二|两|三|四|五|六|七|八|九|十|十一){0,2}个月'

# 中文数字到阿拉伯数字的映射
chinese_to_arabic = {
    '一': 1, '二': 2, '两': 2, '三': 3, '四': 4, '五': 5, '六': 6, '七': 7, '八': 8, '九': 9, '十': 10,
    '十一': 11,  # 添加了“十一”的映射
    # 添加更多的映射如果需要
}



# 解析和转换匹配的句子长度
sentence_lengths = []
for text in reduce:
    # 确保 text 是一个字符串
    if isinstance(text, str):
        matches = re.findall(pattern, text)
        if matches:  # 如果找到匹配项
            for match in matches:
                years, num1, num2 = match
                total_months = 0
                if years:  # 如果有“年”的部分
                    total_months += chinese_to_arabic[years[0]] * 12-1
                if num1:  # 如果有第一个数字
                    total_months += chinese_to_arabic.get(num1, 0)
                if num2:  # 如果有第二个数字
                    total_months += chinese_to_arabic.get(num2, 0)
                sentence_lengths.append(total_months)
        else:  # 如果没有找到匹配项
            sentence_lengths.append(' ')  # 添加空格作为占位符

print(sentence_lengths[:100])  # 打印前100个结果


```

```python
with open("./proed_data/sentence_lengths.csv", "w", encoding="utf-8") as f:
    lines = ["Id,sentence_lengths\n"]
    for i, length in enumerate(sentence_lengths, 1):  # 使用enumerate来获取索引和值
        # 检查length是否是数字，如果不是，用0代替
        length = length if isinstance(length, int) else 0
        lines.append("{},{}\n".format(i, float(length)))
    f.writelines(lines)
```

```python
def modify_list(lst):
    # Iterate through each item in the list
    for i, item in enumerate(lst):
        # Check if the item is a number (integer)
        if isinstance(item, int):
            # Replace the number with the string "明确减刑" followed by the number
            lst[i] = f"明确减刑{item}"
        elif isinstance(item, str) and item.strip() == '':
            # Replace empty strings or strings with only spaces with "未明确"
            lst[i] = "未明确"
    return lst
modified_list = modify_list(sentence_lengths)
```

```python

texts = [t.replace(t[2:t.find('，')],"") for t in texts]
print(texts[:100])
```

```python
#拆分成两部分

for i in range(len(texts)):
    npos = texts[i].find('，')
    name = texts[i][2:npos]
    if len(name) > 1 and len(name) < 5:
        texts[i].replace(name,"")
splits_dicts = ["审理终结", "进行了审理", "进行了审核","审理查明"]
text_pre = []
text_suf = []
for text in texts:
    found_match = False
    for spd in splits_dicts:
        pos = text.find(spd)
        if pos > 0:
            text_pre.append(text[:pos])
            text_suf.append(text[pos:])
            found_match = True  # 找到匹配，更新标志
            break
    if not found_match:
        text_pre.append(text)
        text_suf.append(text)
        
            
            
```

```python

#罪名提取
crime_features = []

cpa1 = '(以|认定|因|对)(被告.{2,5}|被告人|罪犯.{2,5}|罪犯|被告人因|被告人.{2,5}|该犯|其|被告|上诉人（原审被告人）)?(犯|构成)((.{2,15}?罪?、)*(.{2,15}?罪))'
cpa2 = '(以|认定|因|对)((.{2,15}?罪、)*(.{2,15}?罪))判处'
cpa3 = '(被告.{2,5}|被告人|罪犯.{2,5}|罪犯|被告人因|被告人.{2,5}|该犯|其|被告)(犯|构成)((.{2,15}?罪、)*(.{2,15}?罪))'
pattern = cpa1 +'|'+ cpa2 + '|' + cpa3
 
for id,tp in enumerate(text_pre):
    r = re.search(pattern,tp)
    if r:
        if r.group(4):
            crime_features.append(r.group(4))
        elif r.group(8):
            crime_features.append(r.group(8))
        else:
            crime_features.append(r.group(13))
    else:
        crime_features.append(" ")

```

```python
print(len(crime_features))
```

```python

#刑期
sentence_features = []
pa = '(判处|执行)(被告.{2,5}|被告人|罪犯.{2,5}|罪犯|被告人因|被告人.{2,5}|该犯|其|被告)?(有期徒刑.{1,6}年|无期徒刑|死刑)'
patterns = pa
for id,tp in enumerate(text_pre):
    r = re.search(pa,tp)
    if r:
        if r.group(3):
            sentence_features.append(r.group(3))
    else:
        sentence_features.append(' ')

#审查部分处理 切词 删词
stopwords = open("停用词表.txt", encoding="utf-8").read().split('\n')
review_cut = []
for ts in text_suf:
    review_cut.append([token for token in jieba.lcut(ts) if token not in stopwords])
```

```python
crime_features_cut = [cf.split('、') for cf in crime_features]
sentence_features_cut = [jieba.lcut(sf) for sf in sentence_features]
```

```python

datas = []
for cf,sf,rev,sen in zip(crime_features_cut,sentence_features_cut,review_cut,modified_list):
    data = []
    data.extend(cf)
    data.extend(sf)
    data.extend(rev)
    data.append(sen)
    datas.append(data)
```

```python
import csv

input_file = 'proed_data/predict.csv'  # 假设这是你的输入文件
output_file = 'proed_data/output1.csv'  # 这将是包含结果的输出文件

with open(input_file, newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    with open(output_file, 'w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile)
        
        # 读取表头并添加新的 "output" 列
        header = next(reader)
        header.append('output')
        writer.writerow(header)  # 写入新的表头
        
        for row in reader:
            # 转换第二列和第三列为整数
            col2 = int(row[1])
            col3 = int(row[2])
            
            # 应用逻辑处理
            if col3 != 0:
                if abs(col2 - col3) <= 2:
                    result = col2
                else:
                    result = col3
            else:
                result = col2
            
            # 将结果作为 "output" 列添加到行
            row.append(result)
            writer.writerow(row)
```
